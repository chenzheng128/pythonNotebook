````python
import pandas as pd
````

````python
# 读取 excel 原始数据
cnki_df = pd.read_excel('cnki_hotnet_topics.xlsx')
cnki_df
````

````python
# 建立 mysql 会话, 读取测试数据库
from pandas.io import sql

# import cx_Oracle
# ora_conn = cx_Oracle.connect('your_connection_string')
# df_ora = pd.read_sql('select * from user_objects', con=ora_conn)    
# print 'loaded dataframe from Oracle. # Records: ', len(df_ora)
# ora_conn.close()

import MySQLdb

# 创建测试数据与访问权限
"""
CREATE TABLE `hello_mysql` (
  `id` int(11) unsigned NOT NULL AUTO_INCREMENT,
  `name` varchar(20) DEFAULT NULL,
  PRIMARY KEY (`id`)
) ENGINE=InnoDB AUTO_INCREMENT=5 DEFAULT CHARSET=utf8;


INSERT INTO `hello_mysql` (`id`, `name`)
VALUES
	(1,'hello1_mysql'),
	(2,'hello2_mysql'),
	(3,'bbb'),
	(4,'cccc');

grant all privileges on hello.* to test@'%' identified by 'test111';


"""

# 创建连接 
mysql_conn = MySQLdb.connect(host="localhost",user="test", passwd="test111",db="hello")

df_mysql = pd.read_sql('select * from hello_mysql;', con=mysql_conn)    
print 'loaded dataframe from MySQL. records:', len(df_mysql)

# 读取测试数据
df_mysql
````

## to_sql 方法
首先建立表格, 如果遇到 data too long 的错误, 可以考虑修改表结构, 将字段长度增大(默认为63). 

````python
# 导入 cnki 数据
sql.to_sql(cnki_df, con=mysql_conn, name='cnki_df', 
                if_exists='append', flavor='mysql')  #  if_exists='replace'
````

````python
# 查看 cnki 数据
cnki_df_from_db = pd.read_sql('select * from cnki_df;', con=mysql_conn)  
print 'loaded dataframe from MySQL. records:', len(cnki_df_from_db)
````
