### Sympy绘图简介
Source: http://docs.sympy.org/latest/modules/plotting.html#introduction

The plotting module has the following functions: Sympy的plot类有以下函数

* plot: Plots 2D line plots. 2D绘图
* plot_parametric: Plots 2D parametric plots. 2D绘图(带参数)
* plot_implicit: Plots 2D implicit and region plots. 隐式绘图
* plot3d: Plots 3D plots of functions in two variables.  3D绘图
* plot3d_parametric_line: Plots 3D line plots, defined by a parameter. 3D绘图(含参数) 
* plot3d_parametric_surface: Plots 3D parametric surface plots. 3D平面绘图

###plot 绘图函数参考 
http://docs.sympy.org/latest/modules/plotting.html#plotting-function-reference

````python
>>> from sympy import symbols
>>> from sympy.plotting import plot
>>> x = symbols('x')
>>> p1 = plot(x*x)
>>> p2 = plot(x)
>>> p1.append(p2[0])
````

````python
plot(x**2, (x, -5, 5))
````

````python
plot(x, x**2, x**3, (x, -5, 5))
````

````python
plot((x**2, (x, -6, 6)), (x, (x, -5, 5)))
````


###2D绘图(含参数)
sympy.plotting.plot.plot_parametric(*args, **kwargs)[source]

Plots a 2D parametric plot.


````python
>>> from sympy import symbols, cos, sin
>>> from sympy.plotting import plot_parametric
>>> u = symbols('u')
````

````python
plot_parametric(cos(u), sin(u), (u, -5, 5))
````

````python
plot_parametric((cos(u), sin(u)), (u, cos(u)))
````

###3d绘图
sympy.plotting.plot.plot3d(*args, **kwargs)[source]

Plots a 3D surface plot.


````python
>>> from sympy import symbols
>>> from sympy.plotting import plot3d
>>> x, y = symbols('x y')
````

````python
>>> plot3d(x*y, (x, -5, 5), (y, -5, 5))
````

### 3d line 图
sympy.plotting.plot.plot3d_parametric_line(*args, **kwargs)[source]

Plots a 3D parametric line plot.

````python
>>> from sympy import symbols, cos, sin
>>> from sympy.plotting import plot3d_parametric_line
>>> u = symbols('u')
>>> plot3d_parametric_line(cos(u), sin(u), u, (u, -5, 5))
````

###3d surface 图
```
sympy.plotting.plot.plot3d_parametric_surface(*args, **kwargs)
```
Plots a 3D parametric surface plot.

````python
>>> from sympy import symbols, cos, sin
>>> from sympy.plotting import plot3d_parametric_surface
>>> u, v = symbols('u v')
>>> plot3d_parametric_surface(cos(u + v), sin(u - v), u - v,
...     (u, -5, 5), (v, -5, 5))
````

###  implicit equations 隐式不等式(这翻译对吗?)
```
sympy.plotting.plot_implicit.plot_implicit(expr, x_var=None, y_var=None, **kwargs)[source]
```

A plot function to plot implicit equations / inequalities.


````python
>>> from sympy import plot_implicit, cos, sin, symbols, Eq, And
>>> x, y = symbols('x y')
#Without any ranges for the symbols in the expression

#Run code block in SymPy Live
>>> p1 = plot_implicit(Eq(x**2 + y**2, 5))
#With the range for the symbols

#Run code block in SymPy Live
>>> p2 = plot_implicit(Eq(x**2 + y**2, 3),
...         (x, -3, 3), (y, -3, 3))
````
